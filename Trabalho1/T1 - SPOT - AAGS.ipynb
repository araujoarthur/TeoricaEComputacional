```julia
#Work Smarter, Not Harder kkkkkkkkkkkkkkkkkkkk

storedC = []
for n ∈ 0:10
  
  if n >= 2
    enhancedRelation = string(storedC[n-1]," / ", n+2 , "*", n+1)
    println("[ENC] n:", n, " => C[",n+2,"] = ",enhancedRelation)
    push!(storedC, enhancedRelation)
  else
    recurrenceRelation = string( "C[",n,"] / ", n+2 , "*", n+1)
    println("[REC] n:", n, " =>  C[",n+2,"] =",recurrenceRelation);
    push!(storedC,recurrenceRelation)
  end
  
end
```

```julia
c0 = 1
c1 = 1
_solution(x) = (c0 * cosh(x)) + (c1 * sinh(x))
```

```julia
?plot
```

```julia
c0 = 1
c1 = 1
plot(_solution, -2*pi:2*pi, label = "-2π ≤ x ≤ 2π", line=(:black, 0.5, 3, :dot), title = "Grafico da Solução - Itens 3.A, 3.B, 3.C", ylabel ="y(x)", xlabel="x")
plot!(_solution, -pi:pi, line=(:red, 0.5, 3, :dash), label = "-π ≤ x ≤ π")
plot!(_solution, -1:1, line=(:green, 0.5, 6, :solid), label = "-1 ≤ x ≤ 1",)  
savefig("PLOT - A B C.png")
```

```julia
c1 = 2
c0 = 2*c1
plot(_solution, -2*pi:2*pi, label = "-2π ≤ x ≤ 2π", line=(:black, 0.5, 3, :dot), title = "Grafico da Solução - Itens  Item 3.D.I", ylabel ="y(x)", xlabel="x")
savefig("PLOT - D I.png")
```
